```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup as bs
```

```python
url2 = 'https://nsga-results.fusesport.com/competitions.asp?compID=341959&id=485'
browser.visit(url2)
```

```python
html = browser.html
soup = bs(html, 'html.parser')
```

```python
genders = []
ages = []
links = []

gender = "Male"
table_h = soup.find_all('table')
for table in table_h: 
    rows = table.find_all('tr')
    for row in rows:
        cells = row.find_all('td')
        age = cells[0].text
        link = cells[1].a['href']
        
        genders.append(gender)
        ages.append(age)
        links.append(link)
        
    gender = 'Female'
        
```

```python
link_data = pd.DataFrame()
link_data['gender'] = genders
link_data['age'] = ages
link_data['link'] = ['https://nsga-results.fusesport.com/'+ link for link in links]
```
